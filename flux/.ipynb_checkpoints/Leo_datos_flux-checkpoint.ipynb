```python
import numpy as np
import os

from datetime import datetime

path='/home/salio/curso_python2/wpc-original/flux'

listing = os.listdir(path)


#bashCommand =""
#p=os.system(bashCommand)

#for files in listing:
#    if files.endswith('.ghg'):


no=0

for files in listing:
    if files.endswith('.data'):
        no+=1
        mm=str(no).zfill(4)
        salida="marchi_"+mm+".dat"                      

        f=open(files,'r')
        f2=open(salida,'w')
        for i in f:
            linea=f.readline()
            if linea[0] == "DATA":
                TOA5=linea[3]+' '+linea[4]
                dd=datetime.strptime(TOA5, '%Y-%m-%d %H:%M:%S:%f')
                TOA6=datetime.strftime(dd, '"%Y-%m-%d %H:%M:%S.%f"')
                CO2=linea[7]
                H2O=linea[8]
                temp=linea[9]
                u=linea[11]
                v=linea[12]
                w=linea[13]
                diagLI=linea[2]
               
                salida=TOA6+","+u+","+v+","+w+","+temp+","+CO2+","+H2O+','+diagLI+'\n'
                f2.write(salida)
        f.close( )
        f2.close( )
```
